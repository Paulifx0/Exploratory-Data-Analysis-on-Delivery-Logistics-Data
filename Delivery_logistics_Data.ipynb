```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')
df = pd.read_csv("C:/Users/delivery_logistics/Delivery_Logistics.csv")
```

```python
df.head()
```

```python
#Check number of rows, columns, datatypes
df.info()
```

```python
#Check summary statistics (numerical columns)
df.describe()
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
#Check for Blank Strings (looks like data but empty)
(df == "").sum()
```

```python
#Look at Each Column One by One
#This tells you how many deliveries each company handled.
df['delivery_partner'].value_counts()
```

```python
#Delivery Partner Chart
plt.figure(figsize=(10,5))
df['delivery_partner'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Delivery Partner Distribution", fontsize=15)
plt.xlabel("Delivery Partner")
plt.ylabel("Number of Deliveries")
plt.xticks(rotation=45)
plt.show()
```

```python
#Vehicle Type Chart
plt.figure(figsize = (10,5))
df['vehicle_type'].value_counts().plot(kind = 'bar', color = 'coral')
plt.title("Vehicle Usage Distribution", fontsize = 15)
plt.xlabel("Vehicle type")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Delivery mode Distribution
plt.figure(figsize=(8,5))
df['delivery_mode'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Delivery Modes", fontsize=15)
plt.xlabel("Mode")
plt.ylabel("Number of Deliveries")
plt.xticks(rotation=0)
plt.show()
```

```python
#Distance distribution
plt.figure(figsize=(8,5))
plt.hist(df['distance_km'], bins=20, color='purple')
plt.title("Distribution of Delivery Distance", fontsize=15)
plt.xlabel("Distance (km)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['delivery_cost'], bins=20, color='blue')
plt.title("Delivery Cost Distribution", fontsize=15)
plt.xlabel("Delivery Cost")
plt.ylabel("Frequency")
plt.show()
```

```python
#Distance vs cost
#Does cost increase when distance increases?
plt.figure(figsize=(8,5))
plt.scatter(df['distance_km'], df['delivery_cost'], alpha=0.3)
plt.title("Distance vs Delivery Cost", fontsize=15)
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Cost")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.head()
```

```python
corr = numeric_df.corr()
```

```python
#Correlation
#Correlation means:
#How strongly are two things connected?
#Example:
#If distance goes up, does cost also go up?
#If weight increases, does rating change?
#Does weather affect cost?
#We check this using:
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap", fontsize=15)
plt.show()
```
